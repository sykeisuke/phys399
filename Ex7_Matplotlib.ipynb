# Exercise 7: Plotting with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Basic line plot ---
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Function")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.show()
```

```python
# --- 2. Multiple curves ---
y_cos = np.cos(x)

plt.plot(x, y, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")
plt.title("Sine and Cosine")
plt.xlabel("x (radians)")
plt.ylabel("Function value")
plt.legend()
plt.show()
```

```python
# --- 3. Scatter plot ---
x_data = np.random.rand(50)
y_data = np.random.rand(50)

plt.scatter(x_data, y_data)
plt.title("Random Scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# --- 4. Histogram ---
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Gaussian samples")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# --- 5. Mini Exercise ---
# Plot kinetic energy KE = 0.5*m*v^2 for v in range(0, 10)

v = np.linspace(0, 10, 100)
m = 1.0
KE = 0.5 * m * v**2

plt.plot(v, KE)
plt.title("Kinetic Energy vs Velocity")
plt.xlabel("Velocity (m/s)")
plt.ylabel("Kinetic Energy (J)")
plt.show()
```

```python
# --- 6. Mini Exercise 2: Estimate π using Monte Carlo ---

N = 10000
x = np.random.rand(N)
y = np.random.rand(N)

inside = x**2 + y**2 <= 1.0
pi_estimate = 4.0 * np.sum(inside) / N

plt.figure(figsize=(5,5))
plt.scatter(x[inside], y[inside], color="blue", s=1, label="Inside")
plt.scatter(x[~inside], y[~inside], color="red", s=1, label="Outside")
plt.title(f"Monte Carlo π Estimate: {pi_estimate:.4f}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
